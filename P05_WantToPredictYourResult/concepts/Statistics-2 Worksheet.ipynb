#### __Create a normal distribution of mean 100 and standard deviation 20 and another of mean 100 and standard deviation 10 can you state some differences?__

```python
import numpy as np
normal_vals1 = np.random.normal(100,20,1000)
normal_vals2 = np.random.normal(100,10,1000)

```

#### __Plot the both curves and observe the difference__ (Good Task)

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
plt.figure(figsize=(15,5)) #<================ This line sets the figure size so use it before every figure plotting
sns.distplot(normal_vals1) #<====== this create a plot (kind of histogram) for the values just put whatever variable you have
```

```python
plt.figure(figsize=(15,5)) #<================ This line sets the figure size so use it before every figure plotting
sns.distplot(normal_vals2) #<====== this create a plot (kind of histogram) for the values just put whatever variable you have
```
