## Probability Distributions

A probability distribution is a mathematical function that provides the probabilities of occurrence of different possible outcomes in an experiment. In more technical terms, the probability distribution is a description of a random phenomenon in terms of the probabilities of events. For instance, if the random variable X is used to denote the outcome of a coin toss ("the experiment"), then the probability distribution of X would take the value 0.5 for X = heads, and 0.5 for X = tails (assuming the coin is fair).

### __How to get values from a normal distribution?__

Yeah videos are cool but what about doing it by my hands. You don't have to wait for long my friends.

```normal_vals = np.random.normal(100,5,1000)```

What is happening?

Let's break it in parts. 

__np.random__ -> is a module which has different ways of creating random values. Trust me it's a cool module


__np.random.normal__ -> It means that we want to create random values from a normal sample -> like we want values pertaining to a normal distribution

__np.random.normal(100,5,1000)__ -> 100 is mean here, 5 stands for standard deviation (sqrt(variance)) and 1000 is the number of values we want

````python
normal_vals = np.random.normal(100,5,1000)
````

````python
normal_vals.mean() #<-------------- Yeah mean is approx 100
````

````python
normal_vals.var()
````

````python
normal_vals.std()
````

````python
plt.figure(figsize=(15,7)) #<================ This line sets the figure size so use it before every figure plotting
sns.distplot(normal_vals) #<====== this create a plot (kind of histogram) for the values just put whatever variable you have 
````

#### __You can see above that most of the values are concentrated towards the mean(100)__
