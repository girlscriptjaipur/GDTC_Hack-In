### Methods to select data from dataframe

```python
df.head()
```

Pandas __DataFrame.loc__ attribute access a group of rows and columns by label(s) or a boolean array in the given DataFrame. loc takes two single/list/range operator separated by ','. The first one indicates the row and the second one indicates columns.

```python
df.loc['Jignesh'] #Selecting by mentioning the row label
```

```python
df.loc['Jignesh','State'] #WE mentioned Row name as Jignesh and Column as State so yeah Gujrat was the answer
```

```python
df.loc['Anandiben','State':'Sex']
```

__This means select row with label Anandiben but select columns from State to Sex (All of them between them and Sex  Column included)__

#### __.iloc()__
__Pandas provide various methods in order to get purely integer based indexing. Like python and numpy, these are 0-based indexing.__

The various access methods are as follows −

An Integer<br>
A list of integers<br>
A range of values

```python
df.head()
```

```python
df.iloc[0] #0 -> Jignesh
```

```python
df.iloc[0,0] #<- 0 is first row and 0 is first column
```

```python
df.iloc[[0]] #Way to present in a more arranged manner closing no of row or column in [] make it a dataframe
```

```python
#Selecting first 5 rows
df.iloc[0:5]
```

```python
#Selecting 5 rows with State Column and Field Column
df.iloc[0:5,0:2] #2 is excluded
```

```python
#Selecting 5 rows with State and Sex Column
df.iloc[0:5,[0,2]]
```
