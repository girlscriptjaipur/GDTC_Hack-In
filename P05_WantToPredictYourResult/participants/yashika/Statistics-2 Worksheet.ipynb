#### __Create a normal distribution of mean 100 and standard deviation 20 and another of mean 100 and standard deviation 10 can you state some differences?__

```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
normal1=np.random.normal(100,20,1000)
normal2=np.random.normal(100,10,1000)
print(normal1.mean(),normal1.var(),normal1.std())
print(normal2.mean(),normal2.var(),normal2.std())

 
```

#### __Plot the both curves and observe the difference__ (Good Task)

```python
plt.figure(figsize=(15,5))
sb.distplot(normal1)
sb.distplot(normal2)
```
