```python
#solution
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
df_sem_1 = pd.read_csv("firstsem_cse.csv")
df_sem_2 = pd.read_csv("secondsem_cse.csv")
df_sem_3 = pd.read_csv("thirdsem_cse.csv")
df_sem_2.drop(["Total Credits","Total Marks"],axis=1,inplace=True)
merged_inner1 = pd.merge(left=df_sem_1,right=df_sem_2, left_on='EnrollMent No', right_on='Enrollment No')
merged_inner1
merged_inner1.drop(["SID_x","Scheme ID_x"],axis=1,inplace=True)
df_sem_3.drop(["Total Credits","Total Marks"],axis=1,inplace=True)
merged_inner2= pd.merge(left=merged_inner1,right=df_sem_3, left_on='EnrollMent No', right_on='Enrollment No')
#merged_inner2.drop_duplicates()
merged_inner2.set_index(["EnrollMent No"],inplace=True)
merged_inner2.drop(["Total Credits","Student Name_y","Enrollment No_x","SID_y","Scheme ID_y","Student Name","Enrollment No_y"],axis=1,inplace=True)
merged_inner2
```
