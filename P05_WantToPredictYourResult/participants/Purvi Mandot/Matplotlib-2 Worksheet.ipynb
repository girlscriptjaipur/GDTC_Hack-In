```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('https://query.data.world/s/bqrnga7n5nhps5yn5cbiarzdbqv2os')
df.drop(["Unnamed: 8","Unnamed: 9","Unnamed: 10","Unnamed: 11","Unnamed: 12","Unnamed: 13"],axis=1,inplace=True)
df['Party'].unique()
df = df.iloc[:,[0,5,7]]
df = df.groupby(['Party','Financial Year ']).sum() 
df_party['BJP'] = df.loc['BJP','Amount']
df_party['CPI'] = df.loc['CPI','Amount']
df_party['CPI(M)'] = df.loc['CPI(M)','Amount']
df_party['INC'] = df.loc['INC','Amount']
df_party['NCP'] = df.loc['NCP','Amount']
df_party.fillna(value=0,inplace=True)
df_party.plot(kind='area',figsize=(14,7))
plt.ylabel("Donation Amount (Rupees)")
plt.xlabel("Parties")
plt.show() 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel('https://query.data.world/s/bqrnga7n5nhps5yn5cbiarzdbqv2os')
df = df.groupby(['Party','Financial Year ']).sum()
df_party = pd.DataFrame()
df_party['BJP'] = df.loc['BJP','Amount']
df_party['CPI'] = df.loc['CPI','Amount']
df_party['CPI(M)'] = df.loc['CPI(M)','Amount']
df_party['INC'] = df.loc['INC','Amount']
df_party['NCP'] = df.loc['NCP','Amount']
df_party.plot.bar()
plt.show()
```

```python
df_party['Total'] = df_party.sum(axis=1)
plt.hist(df_party['BJP'],bins=11,density=True)
```

```python
df_party['Total'].plot(kind="pie")
plt.title("Total Amount for each party")
plt.show()
```

```python
df_party['Total'].plot(kind="box")
plt.title("Total Amount for each party")
plt.show()
```
