#### __Create a normal distribution of mean 100 and standard deviation 20 and another of mean 100 and standard deviation 10 can you state some differences?__

```python
normal_vals = np.random.normal(100,20,1000)
normal_vals.std()
```

```python
normal_vals2 = np.random.normal(100,10,1000)
normal_vals2.std()
```

#### __Plot the both curves and observe the difference__ (Good Task)

```python
plt.figure(figsize=(15,7))
sns.distplot(normal_vals)
```

```python
plt.figure(figsize=(15,7)) #<================ This line sets the figure size so use it before every figure plotting
sns.distplot(normal_vals2)
```
