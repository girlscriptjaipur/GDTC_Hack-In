
Set The Column Student Name as the Index

```python
df = pd.read_csv("firstsem_cse.csv",index_col="Student Name")
df.head()
```

Drop the Enrollment No, Columns SID,Scheme ID, ALL LAB SUBJECTS (LAB IN THEIR NAME) and Total Credits from the DataFrame df

```python
df = df.drop(['SID','EnrollMent No'],axis=1)
df = df.drop(['Scheme ID','Total Credits'],axis=1)
df.head()
```

```python
df = df[df.columns.drop(list(df.filter(regex='LAB')))]
df.head(
)
```

Add a Column Name Total Marks which will have total marks of each row (Summation of all marks obtained)

```python

df.shape
```

```python
df['Total Marks'] = df.sum(axis=1)
df.head()
```

Combine all the columns
Combine Columns of External and Internal Marks into one i.e AM-1 Internal and AM-1 External must be added and stored to another column named AM-1 Total(drop the AM-1 Internal and AM-1 External After that). Use any of the technique you like and do the same for all the subjects.
ATTENTION

A subject named HVPE doesn't have internal marks.
Dataframe after this step must look like

| Student Name |AM-1 Total|AM-1 Total|AP-1 Total|MP Total|ET Total|HVPE Total| FOC Total|AC Total|Total Marks| |------|------|------|------|------|------|------|------|------|--|

```python
df['AM-1 Total']=df['AM-1 Internal']+df['AM-1 External']

df = df.drop(['AM-1 Internal','AM-1 External'],axis=1)
df['AP-1 Total']=df['AP-1 Internal']+df['AP-1 External']

df = df.drop(['AP-1 Internal','AP-1 External'],axis=1)
df['MP Total']=df['MP Internal']+df['MP External']

df = df.drop(['MP Internal','MP External'],axis=1)

df['ET Total']=df['ET Internal']+df['ET External']

df = df.drop(['ET Internal','ET External'],axis=1)
df['FOC Total']=df['FOC Internal']+df['FOC External']

df = df.drop(['FOC Internal','FOC External'],axis=1)
df['AC Total']=df['AC Internal']+df['AC External']

df = df.drop(['AC Internal','AC External'],axis=1)
df['HVPE Total']=df['HVPE External']

df = df.drop(['HVPE External'],axis=1)
df.head()
```
