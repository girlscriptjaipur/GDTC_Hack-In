### Before Jumping into the ML Let's visualize and find some insights

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Partioning first sem Data

```python
marks_frame_first = df_combined.iloc[:,1:14] #No Need of student name and the Lab Marks only using the theory marks
marks_frame_first.head()
```

```python
#Adding one more column Total Marks but this will be total of first sem marks Including the Lab marks
marks_frame_first['Total Marks'] = df_combined.iloc[:,1:26].sum(axis=1) #Upto AC LAB EXTERNAL
marks_frame_first.head()
```

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(35,25), ncols=2, nrows=4) #We will need 8 plots (7 subjects and 1 Total)
#This initializes 8 plots for us and figsize means total screen size it will divide it among them
#there would be 5 rows and 2 plots in each rows
#sns darkgrid is pure asthetics no need just is soothing for eyes
#the line down also makes frame look pretty
plt.subplots_adjust(left=0.125,bottom=0.1,right=0.8,top=0.9,wspace=.2,hspace=.2)
#Next Line sets title for each frame
ax[0][0].set_title("AM-1",fontsize=30)
ax[0][1].set_title("AP-1",fontsize=30)
ax[1][0].set_title("MP",fontsize=30)
ax[1][1].set_title("ET",fontsize=30)
ax[2][0].set_title("HVPE",fontsize=30)
ax[2][1].set_title("FOC",fontsize=30)
ax[3][0].set_title("AC",fontsize=30)
ax[3][1].set_title("Total Marks",fontsize=30)
sns.distplot(marks_frame_first['AM-1 Internal']+marks_frame_first['AM-1 External'],ax=ax[0][0])
sns.distplot(marks_frame_first['AP-1 Internal']+marks_frame_first['AP-1 External'],ax=ax[0][1])
sns.distplot(marks_frame_first['MP Internal']+marks_frame_first['MP External'],ax=ax[1][0])
sns.distplot(marks_frame_first['ET Internal']+marks_frame_first['ET External'],ax=ax[1][1])
sns.distplot(marks_frame_first['HVPE External'],ax=ax[2][0])
sns.distplot(marks_frame_first['FOC Internal']+marks_frame_first['FOC External'],ax=ax[2][1])
sns.distplot(marks_frame_first['AC Internal']+marks_frame_first['AC External'],ax=ax[3][0])
sns.distplot(marks_frame_first['Total Marks'],ax=ax[3][1])
```
