#### __What do one standard deviation means in normal distribution? Explain using code or way you prefer?__

###### Standard Deviation gives us an idea about the spread of data over the mean. So one standard deviation means mean+std.deviation or mean-std.deviation

#### __Create a normal distribution of mean 100 and standard deviation 20 and another of mean 100 and standard deviation 10 can you state some differences?__

```python
import numpy as np
normal_vals1 = np.random.normal(100,20,1000)
normal_vals2 = np.random.normal(100,10,1000)
print(normal_vals1)
print(normal_vals2)
```

#### __Plot the both curves and observe the difference__ (Good Task)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,7))
sns.distplot(normal_vals1,color="red",label="A")
sns.distplot(normal_vals2,color="Blue",label="B")
```

```python
#Uniform Distribution
data = np.random.uniform(0,1,1000) # Generating 1000 points between 0 and 1.
plt.figure(figsize=(10,5))
sns.distplot(data) 
```

```python
#Binomial Distribution
data = np.random.binomial(10,0.5,1000) # result of flipping a coin 10 times, tested 1000 times.
plt.figure(figsize=(10,5))
sns.distplot(data) 
```

```python
#Poisson Distribution
data = np.random.poisson(5,10000) 
plt.figure(figsize=(10,5))
sns.distplot(data) 
```
